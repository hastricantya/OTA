```python
# Assuming you have a column named 'text' that contains text data
text_column = 'FINAL_FIX'

# Concatenate the text from all rows into a single string
text = " ".join(df[text_column])
```

```python
# Create a WordCloud object with optional customization
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
```

```python
# Display the word cloud using Matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Turn off axis labels and ticks
plt.title('Word Cloud')
plt.show()
```

```python
!pip install pandas matplotlib wordcloud

import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Read the CSV file
df = pd.read_csv("https://raw.githubusercontent.com/hastricantya/OTA/main/FinalFix_Data_OTA.csv")

# Check the columns in the dataset
print(df.columns)

# Filter and separate positive and negative reviews
positive_reviews = df[df['label'] == 'Positive']['FINAL_FIX']
negative_reviews = df[df['label'] == 'Negative']['FINAL_FIX']

def show_wordcloud(text, title):
    # Check if there is any text to generate a word cloud
    if text.strip():  # This checks if the text is not empty or consists of only whitespace
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
        plt.figure(figsize=(10, 5))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(title)
        plt.show()

# Generate and display word clouds for positive and negative reviews
show_wordcloud(" ".join(positive_reviews), 'Word Cloud - Positive')
show_wordcloud(" ".join(negative_reviews), 'Word Cloud - Negative')
```
